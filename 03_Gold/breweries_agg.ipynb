#### Gold Layer
Gold Layer: Create an aggregated view with the quantity of breweries per type and location

```python
from pyspark.sql.functions import *
```

```python
df_silver = spark.read.table("ab_inbev_lake.silver.breweries")
```

```python
df_gold = (
    df_silver
    .groupBy("ds_brewery_type", "nm_country", "nm_state_province")
    .agg(count("*").alias("qty_breweries"))
)
```

```python
df_gold.write.format("delta").mode("overwrite").saveAsTable("ab_inbev_lake.gold.breweries_agg")
```
